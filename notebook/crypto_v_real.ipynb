```python
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce

%matplotlib inline
```

```python
#sourcing data, will import when all neded data located.
btc_val = pd.read_csv('../data/btc_usd.csv')
btc_val.head()
```

```python
btc_val.value_counts()
```

```python
#df['new_date'] = pd.to_datetime(df['old_date'].astype(str))
btc_val['Date'] = pd.to_datetime(btc_val['Date'].astype(str))
btc_val.value_counts()
```

```python
start_date = btc_val.Date.min()
#will need this to remove all date before it from other datasets.
start_date
```

```python
crypto_index = pd.read_csv('../data/cmc200.csv')
oil = pd.read_csv('../data/crude_oil.csv')
dow  = pd.read_csv('../data/dji.csv')
euro  = pd.read_csv('../data/eur_usd.csv')
gbp  = pd.read_csv('../data/gbp_usd.csv')
gold = pd.read_csv('../data/gold.csv')
yen  = pd.read_csv('../data/jpy_usd.csv')
nasdaq  = pd.read_csv('../data/nasdaq.csv')
silver  = pd.read_csv('../data/silver.csv')
sp500  = pd.read_csv('../data/sp500.csv')
```

```python
#data is read in, lets filter it to not have any data before start_date
#before I combine there needs to be a Market column and value added.
btc_val['Market'] = 'Bitcoin'
crypto_index['Market'] = 'CMC Crypto 200'
oil['Market'] = 'Oil'
dow['Market'] = 'Dow Jones Industrial'
euro['Market'] = 'EUR/USD'
gbp['Market'] = 'GBP/USD'
gold['Market'] = 'Gold'
yen['Market'] = 'JPY/USD'
nasdaq['Market'] = 'NASDAQ'
silver['Market'] = 'Silver'
sp500['Market'] = 'S&P 500'
btc_val.head()
```

```python
#need to move Market to front of list. all tables have the same columns so lets do that in a list.
order = ['Market', 'Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
btc_val = btc_val[order]
btc_val.head()
```

```python
crypto_index = crypto_index[order]
oil = oil[order]
dow = dow[order]
euro = euro[order]
gbp = gbp[order]
gold = gold[order]
yen = yen[order]
nasdaq = nasdaq[order]
silver = silver[order]
sp500 = sp500[order]
```

```python
#now to prep to mash it all into a single df
#df_final = reduce(lambda left,right: pd.merge(left,right,on='name'), dfs)

dfs = [btc_val, crypto_index, oil, dow, euro, gbp, gold, yen, nasdaq, silver, sp500]
big_table = pd.concat(dfs).reset_index(drop = True)
big_table.head()
```

```python
big_table['Date'] = pd.to_datetime(big_table['Date'].astype(str))
big_table.tail()
```

```python
big_table = big_table[big_table['Date'] >= start_date].reset_index(drop = True)
big_table.tail()
```

```python
big_table.to_csv('../data/agg_markets.csv')
```
